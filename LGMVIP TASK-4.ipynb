# Importing Libraries

```python
import cv2
from matplotlib import pyplot as plt
```

# Importing Image

```python
image = cv2.imread(r'C:\Users\Lenovo\OneDrive\Desktop\horse.jpg')
```

```python
image
```

# Original Image

```python
plt.imshow(image)
plt.axis(False)
plt.show()
```

# Convert To Gray Image

```python
grayImage = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.axis(False)
plt.imshow(grayImage)
plt.show()
```

# Convert To Invert Image

```python
invertImage = cv2.bitwise_not(grayImage)
plt.axis(False)
plt.imshow(invertImage)
plt.show()
```

# Convert To Smooth Image

```python
image_smoothing = cv2.GaussianBlur(invertImage, (21,21),sigmaX=0, sigmaY=0)
plt.axis(False)
plt.imshow(image_smoothing)
plt.show()
```

```python
def dodgeV2(x,y):
    return cv2.divide(x,255-y, scale=256)
```

# Pencil Sketch

```python
Sketch_image = dodgeV2(grayImage,image_smoothing)
plt.axis(False)
plt.imshow(Sketch_image)
plt.show()
```
